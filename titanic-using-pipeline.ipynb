```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
df
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
# Step 1 -> train/test/split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['Survived']),
                                                 df['Survived'],
                                                 test_size=0.2,
                                                random_state=42)
```

```python
X_train.head()
```

```python
y_train.sample(5)
```

# Blocks for PipeLine

```python
# imputation transformer
trf1 = ColumnTransformer([
    ('impute_age',SimpleImputer(strategy='mean'),[2]),
    ('imputer_embarked',SimpleImputer(strategy='most_frequent'),[6])
],remainder='passthrough')
```

```python
# one hot encoding
trf2 = ColumnTransformer([
    ('ohe_sex',OneHotEncoder(sparse_output=False, handle_unknown='ignore'),[1]),
    ('ohe_embarked',OneHotEncoder(sparse_output=False, handle_unknown='ignore'),[6])
],remainder='passthrough')
```

```python
# scaling
trf3 = ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,10))
])
```

```python
# # Feature selection
# trf4 = SelectKBest(score_func=chi2,k=8)
```

```python
# train the model
trf5 = DecisionTreeClassifier()
```

# Create Pipeline

```python
pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),
    # ('trf4',trf4),
    ('trf5',trf5)
])
```

```python
pipe.fit(X_train,y_train)
```

```python
pipe.named_steps['trf1'].transformers_[0][1].statistics_
```

```python
pipe.named_steps
```

```python
from sklearn import set_config
set_config(display='diagram')
```

```python
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(pipe, X_train, y_train, cv=5, scoring='accuracy').mean()
```
