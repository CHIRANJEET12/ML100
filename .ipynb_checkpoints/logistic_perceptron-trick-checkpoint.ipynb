```python
def perceptron(X,y):
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1

    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]

    return weights[0],weights[1:]
    
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_,coef_ = perceptron(X,y)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python

# 5 samples, 2 features
X = np.array([
    [2, 3],
    [1, 5],
    [2, 1],
    [3, 4],
    [5, 2]
])
y = np.array([1, 1, 0, 1, 0])   # binary labels
```

```python
def step(z):
    return 1 if z >= 0 else 0
```

```python
def perceptron_batch(X, y, lr=0.1, epochs=10):
    X = np.insert(X, 0, 1, axis=1)   # add bias
    weights = np.ones(X.shape[1])

    for _ in range(epochs):
        for j in range(X.shape[0]):   # loop through ALL samples
            y_hat = step(np.dot(X[j], weights))
            weights = weights + lr * (y[j] - y_hat) * X[j]

    return weights
```

```python
print("Batch:", perceptron_batch(X, y))
```

```python
def perceptron_sgd(X, y, lr=0.1, epochs=10):
    X = np.insert(X, 0, 1, axis=1)
    weights = np.ones(X.shape[1])

    for _ in range(epochs * X.shape[0]):   # ~like 1 epoch = 5 updates
        j = np.random.randint(0, X.shape[0])  # pick 1 random sample
        y_hat = step(np.dot(X[j], weights))
        weights = weights + lr * (y[j] - y_hat) * X[j]

    return weights
```

```python
print("SGD:", perceptron_sgd(X, y))
```

```python
def perceptron_minibatch(X, y, lr=0.1, epochs=10, batch_size=2):
    X = np.insert(X, 0, 1, axis=1)
    weights = np.ones(X.shape[1])

    for _ in range(epochs):
        idx = np.random.choice(X.shape[0], batch_size, replace=False)
        X_batch, y_batch = X[idx], y[idx]
        
        for j in range(batch_size):
            y_hat = step(np.dot(X_batch[j], weights))
            weights = weights + lr * (y_batch[j] - y_hat) * X_batch[j]

    return weights
```

```python
print("Mini-Batch:", perceptron_minibatch(X, y))
```
