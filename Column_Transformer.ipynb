```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['has_covid']),df['has_covid'],
                                                test_size=0.2)
```

# 1. Aam Zindagi

```python
# fill in the missing values in fever
si = SimpleImputer(strategy="mean")
X_train_fever = si.fit_transform(X_train[['fever']])

X_test_fever = si.transform(X_test[['fever']])

X_train_fever.shape
```

```python
# Ordinalencoding in cough
oe = OrdinalEncoder(categories=[['Mild','Strong']])
oe
```

```python
X_train_oe = oe.fit_transform(X_train[['cough']])
X_tst_oe = oe.transform(X_test[['cough']])
```

```python
X_train_oe.shape
```

```python
# OneHotEncoding in gender,city
ohe = OneHotEncoder(drop='first',sparse_output=False)
X_train_ohe = ohe.fit_transform(X_train[['gender','city']])
X_test_ohe = ohe.transform(X_test[['gender','city']])
```

```python
X_train_ohe.shape
```

```python
# Extracting Age
X_train_age = X_train.drop(columns=['gender','fever','cough','city']).values
X_test_age = X_test.drop(columns=['gender','fever','cough','city']).values

X_train_age.shape
```

```python
X_train_transformed = np.concatenate((X_train_age,X_train_ohe,X_train_oe,X_train_fever),axis=1)
X_test_transformed = np.concatenate((X_test_age,X_test_ohe,X_tst_oe,X_test_fever),axis=1)

X_train_transformed
```

# Mentos Zindagi

```python
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('covid_toy.csv')
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['has_covid']),df['has_covid'],
                                                test_size=0.2)
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['gender','city'])
],remainder='passthrough')
```

```python
transformer.fit_transform(X_train)
```

```python
transformer.transform(X_test)
```
